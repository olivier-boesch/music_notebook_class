Importation

```python
from music import Music
```

cr√©ation d'une note

```python
la3 = Music(440)
```

Affichage du lecteur

```python
la3.play()
```

Affichage des donn√©es sous forme d'un graphique

```python
la3.plot(periods=5)
```

Changement de fr√©quence

```python
demi_ton = 2**(1/12)  # d√©finition du demi ton en temp√©rament √©gal

do4 = la3 * demi_ton**3

do4.play()
```

Cr√©ation d'un son complexe

```python
tierce_mineure = la3 & do4

tierce_mineure.play()
```

Affichage du son complexe

```python
tierce_mineure.plot(periods=15)
```

Cr√©ation de m√©lodie



```python
tempo = 1.5  # 90bpm = 1.5 temps par seconde
noire = 1/tempo
blanche = 2 * noire

freq_do4 = 440 * demi_ton**3
freq_re4 = freq_do4 * demi_ton**2
freq_mi4 = freq_re4 * demi_ton**2
clair_de_la_lune = Music(freq_do4, noire) + Music(freq_do4, noire) + Music(freq_do4, noire)
clair_de_la_lune += Music(freq_re4, noire) + Music(freq_mi4, blanche) + Music(freq_re4, blanche)
clair_de_la_lune += Music(freq_do4, noire) + Music(freq_mi4, noire) + Music(freq_re4, noire) + Music(freq_re4, noire)
clair_de_la_lune += Music(freq_do4, blanche)
                
clair_de_la_lune.play()
                   
```
